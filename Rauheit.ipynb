```python
import pathlib

import matplotlib.pyplot as pl
import numpy as np
import scipy.fft
```

```python
images = {path.stem: pl.imread(path) for path in pathlib.Path("Papiere").glob("*.png")}
```

```python
images["Canson"].shape
```

```python
ffts = {name: np.abs(scipy.fft.rfft2(image[:, :, 0])) for name, image in images.items()}
```

```python
fig = pl.figure(figsize=(15, 8))
for i, (name, fft) in enumerate(ffts.items(), 1):
    x = np.abs(fft)
    x[0, 0] = 0
    x[0, -1] = 0
    ax = fig.add_subplot(2, 3, i)
    im = ax.imshow(x)
    ax.set_title(name)
    pl.colorbar(im, ax=ax)
pl.tight_layout()
pl.show()
```

```python
fig = pl.figure(figsize=(15, 8))
for i, (name, fft) in enumerate(ffts.items(), 1):
    x = np.abs(fft)
    x[0, 0] = 0
    x[0, -1] = 0
    ax = fig.add_subplot(2, 3, i)
    im = ax.imshow(x[:30, :30])
    ax.set_title(name)
    pl.colorbar(im, ax=ax)
pl.tight_layout()
pl.show()
```

```python
for name, image in images.items():
    pl.clf()
    pl.plot(image[150, :, 0], label=name)
    pl.title(name)
    pl.show()
```

```python
import scipy.signal
```

```python
o = np.ones_like(images['Kopierpapier'][:, :, 0])
corr_norm = scipy.signal.correlate(o, o)
```

```python
pl.clf()
pl.imshow(corr_norm)
pl.colorbar()
pl.title('Correlation Norm')
pl.grid()
pl.show()
```

```python
c1 = scipy.signal.correlate(images['Kopierpapier'][:, :, 0], images['Kopierpapier'][:, :, 0])
c2 = scipy.signal.correlate(images['Pinkes Skizzenbuch'][:, :, 0], images['Pinkes Skizzenbuch'][:, :, 0])
```

```python
pl.clf()
pl.imshow(c1 / corr_norm)
pl.colorbar()
pl.title('Kopierpapier')
pl.grid()
pl.show()
```

```python
pl.clf()
pl.imshow(c2 / corr_norm)
pl.colorbar()
pl.title('Pinkes Skizzenbuch')
pl.grid()
pl.show()
```
